```python
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'LLMs': ['gpt-3.5-turbo', 'gpt-3', 'ada, babbage, curie', 'claude', 'claude-instant', 'command-xlarge', 
              'command-medium', 'BERT', 'T5', 'PaLM', 'LLaMa', 'CTRL', 'Dolly 2.0'],
    'Params (billions)': [175, 175, 7, 52, 52, 50, 6, 0.345, 11, 540, 65, 1.6, 12]
}

plt.figure(figsize=(10, 6)) 

sns.barplot(x='LLMs', y='Params (billions)', data=data)
plt.xticks(rotation=45)

plt.xlabel('LLMs', fontsize=14)
plt.ylabel('Params', fontsize=14)
plt.title('Number of Params for Different LLMs', fontsize=16)

plt.show()
```
